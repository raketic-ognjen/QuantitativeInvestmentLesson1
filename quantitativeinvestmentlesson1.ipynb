## <font color='orange' style="font-size:25px"><b>Real vs Financial Assets</b></font>

**Read Bodie, Kane, Markus, Chapter 1, up to p. 14 for more details.**

The material wealth of a society producted by **real assets**: the land, buildings, machines, and knowledge that can be used to produce goods and services.

**Financial assets** such as stocks and bonds are no more than sheets of paper or, more likely, computer entries. These are the **means** by which individuals in well-developed economies hold claims on real assets.

Financial assets are claims to the income generated by real assets (or claims on income from the government). E.g., unless I am Mate Rimac, it may be hard to own an auto plant (a real asset), but I can buy shares in Tesla (shares are financial asset) and thereby share in the income derived from the production of electric automobiles.

Real assets **generate net income** to the economy, financial assets **allocate income or wealth among investors**.


People select between consuming their wealth today or invest for the future. If they choose to invest, they may place their wealth in financial assets by purchasing various securities.

When investors buy these securities from companies, the firms use the proceeds to pay for real assets, such as plant, equipment, technology, inventory, or, very importantly, human capital.

So investors’ returns on securities ultimately come from the income produced by the real assets that were financed by the issuance of those securities.

## <font color='orange' style="font-size:25px"><b>Financial Assets</b></font>

The **financial markets** provide a myriad of mechanisms for raising funds and managing and trading risks, providing opportunities for investors, fund managers, and speculators.

Investor *E.O. Thorpe* once remarked that the difference between gambling and investment is that the latter sounds so much more dignified.

### <font color='MediumVioletRed' style="font-size:20px"><b>Stocks</b></font>

Common stock or equity in a firm represents an ownership
share in the corporation. Equity holders are not promised any particular payment.There are no guarantees beyond the ‘limited liability’ rules under which the value of a share can never be negative. Equityholders are not promised any particular payment.

Shareholders
receive any dividends the firm may pay and have prorated ownership in the real assets of
the firm. If the firm is successful, the value of equity increases; if not, it will decrease.

The performance of equity investments, therefore, is tied directly to the success of the firm and its real assets. For this reason, equity investments tend to be riskier than investments debt instruments. It is, also, the reason, why shareholders and, say, bondholders, typically play the key role in selecting who manages the firm (well, sort of)

shareholders.jpg

### <font color='MediumVioletRed' style="font-size:20px"><b>Bonds</b></font>

Another way companies can raise money is by issuing bonds. Purchasers of bonds are lending money (the ‘face value’) to the company for a fixed period of time, say five years.

The company pays ‘coupons’, generally annually or semi-annually (every six months) which are interest rate payments at an agreed rate, either at a ‘fixed’ rate, meaning that the rate is fixed at, say, 5 per cent per annum for the lifetime of the bond, or ‘floating’ if the rate follows current market rates in an agreed way.

At maturity the company makes the final interest payment and repays the face value. Once issued, a bond is a tradable asset that can be bought and sold on trading platforms at current market prices.

In issuing bonds the company is committing itself to making the scheduled payments on time and in full, regardless of its commercial success. If it fails to do so its credit rating will suffer, and raising further funds will be more difficult and more expensive.

The purchaser receives **a predictable income stream** while he holds the bond but may be able to profit by selling it at an opportune time when market rates are favourable.

**If the company defaults**, the outcome for the purchaser is highly uncertain—he may receive something not far short of the face value or, in the worst case, nothing at all. In short, the ‘risk profile’ of a bond is very different from that of stocks.

### <font color='MediumVioletRed' style="font-size:20px"><b>Foreign exchange</b></font>

The foreign exchange (‘FX’, ‘Forex’) market is the largest and most liquid market in the world. Unlike the stock and futures market that are housed in central physical exchanges, the FX market is an OTC (‘over-the-counter’) decentralized market housed on electronic trading platforms accessible worldwide.

The **US dollar is involved in 85 per cent** of FX trading volume. The Euro is second with 40 per cent, followed by Japanese yen with around 20 per cent of the trading volume.

With volume concentrated mainly in these three currencies, FX traders can focus their attention on just a handful of major pairs. The volume of trading is truly extraordinary.

In 2016 the average daily volume was about $\$5$tn, roughly 90 per cent of it being generated by currency speculators capitalizing on intra-day price movements.

By way of comparison, the daily volume in futures markets was $\$450$bn, equities $\$200$bn, the NY Stock Exchange $\$30$bn. The annual GDP of the USA in 2016 was about $\$18.6$tn.

### <font color='MediumVioletRed' style="font-size:20px"><b>Forwards, futures, and options</b></font>

Now we come to derivative contracts, or just ‘derivatives’, which are ubiquitous in finance and the focus of much of the academic and practitioner research into correct pricing.

‘Derivative’ means that the value of the contract depends on, or is ‘derived from’, the value of one or more underlying assets such as stock prices or FX rates.

Two typical uses of derivatives is to hedge or transfer risk (in risk management) or to speculate on direction or volatility of particular asset or asset class or the entire market.

## <font color='orange' style="font-size:25px"><b>The Players</b></font>

Three major players in the financial
markets:

1. **Firms** are net borrowers. They raise capital now to pay for investments in plant and equipment. The income generated by those real assets provides the returns to investors who purchase the securities issued by the firm.

2. **Households** are net savers (in aggregate!). They purchase the securities issued by firms that need to raise funds.

3. **Governments** can be net borrowers or lenders

Corporations and governments do not sell most of their securities directly
to individuals. About half of all stock held by large financial institutions
such as pension funds, mutual funds, insurance companies, and banks.

These financial institutions stand between the security issuer (the firm, financial institution, government) and the ultimate owner of the security (investors). For this reason, they are called **financial intermediaries**.

Corporations do not market their own securities to the public. Instead, they hire agents, called **investment bankers**, to represent them to the investing public.

**Investment companies** pool and manage the money of many investors.
Here, the problem is that most household portfolios are not
large enough to be spread across a wide variety of securities. In terms of brokerage fees and research costs, purchasing one or two shares of many different firms is very expensive.

**Mutual funds** have the advantage of large-scale trading and portfolio management, while participating investors are assigned a prorated share of the total funds according to the size
of their investment.

Exposure to funds gives small investors advantages they are willing to pay
for via a management fee to the mutual fund operator. Mutual funds are sold in the retail market, and their investment
philosophies are differentiated mainly by strategies that are likely to attract a large number
of clients.

Actively vs passively managed funds. Actively managed aim to beat the market or some other benchmark. Passive funds aim to closely follow an index.

**ETFs (Exchange-Traded Funds)** provide exposure to a particular index, commodity, and liquidity of large stocks. I.e. they are passively managed funds. One can get indirect exposure to, say, Egyptian stock market through an ETF. The largest provider of indices and ETFs is BlackRock.

Economies of scale also explain the proliferation of **analytic services** available to investors. Newsletters, databases, and brokerage house research services all engage in research to be sold to a large client base.

## <font color='orange' style="font-size:25px"><b>Investment process</b></font>

An investor’s portfolio is investor's collection of investment assets. Once the portfolio is established, investor may periodically

* update or “rebalance” by selling existing securities and using the proceeds to purchase new securities

* increase or decrease the size of the portfolio

Investment assets:  stocks, bonds, real estate, commodities, and so on.

**Top-down** portfolio construction starts with asset allocation.

* Asset allocation (select among broad asset classes). The first decision: how much money to put into safe asset (risk-free) and how much into risky assets. Establishes broad features of a portfolio.

* Security selection (select individual securities within an asset class). Fine-tuning.

Asset allocation choice depends to a large extent on investor preferences i.e. risk aversion (see below). Stocks have much higher expected returns than T-bills, but are much riskier too.

**Security analysis**: valuation of particular securities that might be included in the portfolio. For example, an investor might ask whether Moderna or Pfizer is more attractively priced. Both bonds and stocks must be evaluated for investment attractiveness, but valuation is far more difficult for stocks because a stock’s performance usually is far more
sensitive to the condition of the issuing firm.

**Bottom-up** strategy: pick stocks that seem attractively priced without as much concern for the resultant asset allocation. Such a technique can result in unintended bets on one or another sector of the economy.

For example, it might turn out that the portfolio ends up with a very heavy representation of firms in one industry, from one part of the country, or with exposure to one source of uncertainty. However, a bottom-up strategy does focus the portfolio on the assets that seem to offer the most attractive investment opportunities.

A scientific basis for fund management top-down approach was initiated in 1952 when Harry Markowitz — subsequently, an Economics Nobel Prize laureate—published his paper, ‘Portfolio Selection’, which begins as follows:

"The process of selecting a portfolio may be divided into two stages.

- The first stage starts with observation and experience and ends with beliefs about future performances of available securities.

- The second stage starts with the relevant beliefs of future performance and ends with the choice of portfolio. The paper is concerned with the second stage."

The first stage is far harder, and there is never a definitive solution. Predicting the future is difficult in most situations, but particularly so for asset prices in financial markets, which are notoriously erratic.

Indeed, the ‘efficient markets hypothesis’, accepted by some economists, postulates that all information relevant to systematic prediction is already ‘priced in’ to today’s market, leaving only ‘white noise’ as the future. If true, no active fund management makes any sense.

Traders, on the other hand, seek to take advantage of market anomalies to make arbitrage profits, an enterprise that would be impossible in an efficient market.

The difficulties of the first stage do not mean that nothing can be
done in relation to the second stage. The beliefs about future
performances will include acceptance that the future is grossly
uncertain. In this course we discuss various strategies that mitigate the risks of unfavourable scenarios and take advantage of favourable ones.

Below we develop some simple models that are aimed to give us intuition regarding optimal portfolio allocation from a theoretical perspective. For simplicity we consider just a single period of time and a fixed amount of money that we have for investment.

## <font color='orange' style="font-size:25px"><b>Utility function and risk aversion</b></font>

**To learn more about utility functions and risk aversion you can read, for example, Ang, Chapter 2 (especially pages 40-46).**

Consider a utility function of a typical investor

Untitled.png

Note that utility function is increasing in wealth. Of course, we do not consume money itself. Rather, we use money to purchase with it consumption goods. However, assuming that at one point we convert our terminal wealth into consumption we can talk about utility that money provides to us.

Marginal utility is a change in our utility if we increase our wealth by a small amount (say, 1 dollar). Increasing utility function implies that marginal utility of wealth is always positive $U'(W) > 0$. We all prefer to be richer than to be poorer.

However, when we are very poor, each additional dollar means a lot. But, when we get loaded, each additional dollars means less.

In mathematical terms this implies that marginal utility is a decreasing function of wealth, i.e. that $U''(W) < 0$. Functions which satisfy this condition are concave.

Concavity implies aversion towards risk. A drop of our wealth by 1 dollar reduces our utility by a higher amount than what we gain in terms of utility if we get richer by one dollar. The same is true when we compare gaining or losing 200 (as in the figure).

How can we compare different investments? Suppose we have 1000 dollars. We can enter, initially costlessly, into a project that would give us a 50 percent chance to making extra 200 dollars. In that case we would end up with 1200 dollars.

However, there is also a 50 percent chance that we lose 200 dollars. In that case, we would end up with 800 dollars. If we are risk averse, should we accept this risk or should we just keep our 1000 dollars (no investment)?

To make a comparison we use the notion of the expected utility. In case of undertaking the investment the expected utility would be

$$\mathbb{E}u_1 =\mathbb{E}(u(W)) = 0.5 u(1200)+ 0.5 u(800)$$

Note that the expected value of the first opportunity is

$$\mathbb{E}(W) = 0.5 \times 1200+ 0.5 \times 800 = 1000$$

It turns out that no-risk choice would give us higher ex-ante (expected) utility so we would prefer it. (Explain on the graph)

To make it concrete, suppose that the utility function is the log function, i.e. $u(W) = \log(W)$. In that case

```python
EU_1 = 0.5*np.log(1200)+0.5*np.log(800)
EU_1
```

## <font color='orange' style="font-size:25px"><b>Maximizing expected utility to find optimal portfolio</b></font>

Consider now a highly simplified investment problem. Assume that investor has, initially, amount $W_0$ to invest. For simplicity, there are just 2 assets: a riskless asset and a risky asset (stock or an ETF for example).

Our task is to determine how to split out money between riskless and risky investment in an optimal way, taking into the account our risk aversion or, in other words, our preferences (utility function).

We select portfolio at time 0. At time 1 the investment process ends and we consume our portfolio wealth $W_1$. I.e. we leave nothing for some future consumption or investment. While very simplistic, this models captures some very important features of a realistic investment process.

Our preferences are described by a concave utility function $u(W)$ (concave means that marginal utility is a decreasing function or that second derivative is negative). Since our wealth at time 1 is uncertain, we maximize, at time 0, the **expected value of our utility**:

$$\mathbb{E}u=\mathbb{E}(u(W_1))$$

In performing the optimization investor has to take into the account the budget, regulatory and, possibly, a host of other constraints. In this lecture we focus on budget constraints only.

At time 0, we have $W_0$ euros to invest either into a risk-free asset (with interest $r$) or into risky asset (stock or ETF fund, say). The number of stocks we hold at time 0 is $n_s$. The amount, at time 0, invested in risk-free asset is $B$. Stock price at time 0 is $p_{s0}$. The budget constraint at time 0 reads:

$$W_0 = B + n_s\,p_{s0}$$

The amount invested in stock is $n_s\,p_{s0}$, while the amount invested into riskless asset is $B$. At time 1, a share price is worth $p_{s1}$ (this can include both dividends and capital gains). Thus, at time 1, we face the following budget constraint:

$$W_1 = B(1+r) + n_s\,p_{s1}$$

Solving for $B$ from the first budget constraint and replacing it into the second one yields:

$$B = W_0 - n_s\,p_{s0}$$

$$W_1 = (W_0 - n_s\,p_{s0})(1 + r) + n_s\,p_{s1} =W_0(1 + r)+ n_s (p_{s1} - p_{s0}) - n_s\,p_{s0}\,r$$

Portfolio wealth at time 1 (and, therefore, our consumption at time 1) would be $W_0(1+r)$ if we put all our money into the riskless asset. Otherwise, $W_1$ is a random variable. If $n_s\,(p_{s1}-p_{s0})-n_s p_{s0}\,r > 0$, it would pay to have invested in stock and vice versa. The expected utility is:

$$\mathbb{E}u=\mathbb{E}(u(W_1)) = \mathbb{E}(u(W_0 (1 + r) + n_s (p_{s1} - p_{s0}) - n_s\,p_{s0}\,r))$$

### <font color='MediumVioletRed' style="font-size:20px"><b>A numerical example with log utility function and two states of the nature</b></font>

> Suppose that our initial wealth is $1000$ dollars. We can put our money into either a riskless asset and earn return of $r=3\%$, or into a risky asset. Price of the risky asset can go up or down in the next period by $20\%$. The initial price of risky asset is $50$ dollars. Assume that you have estimated that the probability of the up state is $60\%$ while the probability of the down state is $40\%$. Finally, assume that our utility $u(W) = \log(W)$.

```python
W0 = 1000 #Initial Wealth of the investor
ps0 = 50 # Initial stock price, i.e. stock price at time 0
u = 1.2 #Gross return in up state (1+20 percent return)
d = 0.8 #Gross return in down state (1- 20 percent return)
r = 0.03

prob_nat = sp.Matrix(1,2,[0.6,0.4])
prob_nat
```

Possible stock prices at time 1 are, therefore:

```python
ps1 = np.round([ps0*u,ps0*d])
ps1
```

Here, $u$ and $d$ are gross returns on risky asset.

We can present the evolution of riskless and risky asset using the following tree:

```python
fd.tree(fd.bin_evolution(50,[1.2,0.8],1), fd.bin_evolution(1,[1.03,1.03],1),space=0.25, fs=(500,500))
```

For the fixed number of stocks $n_s$, portfolio wealth has two possible realizations (due to the uncertainty generated by the stock price):

```python
ns= sp.Symbol('n_s') #Number of stocks
W1 = sp.Matrix(W0*(1+r)+ns*(ps1-ps0)-ns*ps0*r) #portfolio wait at time 1
W1
```

The first expression above is portfolio wealth at $t=1$ if the up state occurs, $W_{1u}$, while the second expression represents the realization of wealth in the down state, $W_{1d}$. The expected utility is given by:

$$\mathbb{E}u = 0.6 \log(W_{1u}) + 0.4 \log(W_{1d})$$

In contrast to <font color='DodgerBlue'><b>np.array</b></font>, <font color='DodgerBlue'><b>sp.Matrix</b></font> **are not vectorized**. Thus if we want to apply some function (like <font color='DodgerBlue'><b>sp.log</b></font>) on each element of <font color='DodgerBlue'><b>sp.Matrix</b></font> we need to use method: <font color='DeepPink'><b>applyfunc</b></font>. Method has only one required argument - function that you want to apply:

```python
W1.applyfunc(sp.log) #this is the method calculating log of wealth W1 in sympy
```

Now we have the levels of utility in both states of the nature. To calculate the expected utility we dot-multiply the vector of utility realizations with the vector of probabilities:

```python
EU=(prob_nat@W1.applyfunc(sp.log))[0] #Calculating expected utility corresponding to log utility
EU
```

To find the maximum expected utility $\mathbb{E}u$, we find the first order condition with respect to $n_s$ (the number of shares of risky asset is here the decision variable):

```python
FOC = sp.diff(EU,ns) #Calculating first derivative of the expected utility with return to the number of stock
FOC
```

```python
ns_opt= sp.solve(FOC,ns)[0]
ns_opt
```

Thus, if we have log utility preferences, we would have optimally purchased 5.27 stocks (for the price of 50 dollars each). That means that we would have invested the amount of:

```python
ps0*ns_opt
```

... into risky asset and the amount:

```python
W0 - ps0*ns_opt
```

into riskless asset.

### <font color='MediumVioletRed' style="font-size:20px"><b>Different decision variable: optimal amount of money invested into the risky asset </b></font>

Let us come back to the general expression for wealth at time 1.

$$W_1 = W_0(1 + r)+ n_s (p_{s1} - p_{s0}) - n_s\,p_{s0}\,r$$

If we divide and multiply by $p_{s0}$ the expression $n_s (p_{s1}-p_{s0})$ from the equation above, we obtain:

$$W_1=W_0(1 + r)+ n_s\,p_{s0}\, \frac{(p_{s1} - p_{s0})}{p_{s0}} - n_s\,p_{s0}\,r$$

Simple return on stock is $r_s$:

$$r_s=\frac{(p_{s1} - p_{s0})}{p_{s0}}$$

We now introduce new variable $w_{s0}$ which represents the amount of money invested into risky asset:

$$w_{s0}=n_{s}\,p_{s0}$$

Thus, in this notation $W_1$ reads:

$$W_1=W_0(1 + r)+ w_{s0}\, r_s - w_{s0}\,r = W_0(1 + r) + w_{s0} (r_s - r)$$

$W_1$ differs from the risk-free amount $W_0 (1+r)$ by the product of the amount of money invested in risky stock, $w_{s0}$, and the excess return on risky stock $(r_s-r)$. The expected utility would, thus, read:

$$\mathbb{E}u=\mathbb{E}(u(W_0(1 + r)+ w_{s0} (r_s - r)))$$

Now the decision variable would be the **amount of money invested into the risky asset** $w_{s0}$. Clearly, once we find $w_{s0}$, we could immediately determine the optimal number of risky stocks:

$$n_s=\frac{w_{s0}}{p_{s0}}$$

### <font color='MediumVioletRed' style="font-size:20px"><b>Continuation of the numerical example</b></font>

In order to utilize this form of the portfolio optimization problem, let us find first the returns on risky stock $r_s$. Since gross returns are $u$ and $d$, net returns in the two states of nature are:

```python
# Calculating net stock returns starting from the gross stock returns
rs=np.array([u,d])-1
rs
```

Possible wealth at time 1 now reads:

```python
#Wealth invested into risky asset
ws0 = sp.Symbol('w_s0')
#portfolio wealth at time 1
W1= sp.Matrix(W0*(1+r)+ws0*(rs-r))
W1
```

while the expected utility is now given by:

```python
EU=(prob_nat@W1.applyfunc(sp.log))[0]
EU
```

In order to maximize our utility, we need to invest into the risky asset the following amount of money:

```python
FOC = sp.diff(EU,ws0)
FOC
```

```python
ws0_opt=sp.solve(FOC,ws0)[0]
ws0_opt
```

Not surprisingly, we obtained the same answer as before. This implies that we would need to invest 26.3 per cent of our 1000 euros into stock and the rest into riskless asset.

It turns out that we can obtain optimal fraction directly (the third way of expressing the optimization problem). This is what we do next.

### <font color='MediumVioletRed' style="font-size:20px"><b>Optimization using fraction of money invested into the risky asset</b></font>

Let us continue from where we have left. Let me remind you that we have expressed wealth at time 1 as:

$$W_1=W_0(1 + r)+ w_{s0} (r_s - r)$$

Now we are going to divide and multiply the expression $w_{s0} (r_s - r)$ by the initial wealth $W_0$:

$$W_1=W_0(1 + r)+ W_{0}\,\frac{w_{s0}}{W_{0}} (r_s - r)$$

Let $\alpha_{s0}$ be the fraction of money invested into the risky asset:

$$\alpha_{s0}=\frac{w_{s0}}{W_{0}}$$

The expresion for $W_1$ would now read:

$$W_1=W_0 (1 + r) +W_0\,\alpha_{s0} (r_s - r)=W_0 (1 + r+\alpha_{s0} (r_s - r))$$

Expression above can be rewritten as:

$$W_1=W_0 (1 + (1-\alpha_{s0})r+ \alpha_{s_0}\,r_s)$$

Here, $\alpha_{s0}$ and $(1-\alpha_{s0})$ are portfolio weights (fractions of invested wealth) in stock and riskless asset, respectively. Introducing portfolio return $r_p = (1-\alpha_{s0})r+ \alpha_{s_0}\,r_s$ we obtain:

$$W_1=W_0 (1 + r_p)$$

This expression is easy to generalize for the case of multiple assets. Namely, portfolio return is equal to the weighted average of returns on individual portfolio components. Portfolio wealth is the initial wealth plus the profit made from investment into the portfolio.

Our decision variable is now $\alpha_{s0}$

### <font color='MediumVioletRed' style="font-size:20px"><b>Continuation of the numerical example</b></font>

Now the expression for wealth at time 1 is:

```python
as0 = sp.Symbol('alpha_s0')
W1 = sp.Matrix(W0*(1+r*(1-as0)+as0*rs))
W1
```

While the expected utility is now given by:

```python
EU=(prob_nat@W1.applyfunc(sp.log))[0]
EU
```

Finally, in order to maximize our utility, we need to invest following fraction in risky asset:

```python
FOC = sp.diff(EU,as0)
FOC
```

Suppose we are twice as wealthly as before. Would optimal fraction invested into risky asset change?

```python
W0=2000
W1=sp.Matrix(W0*(1+r*(1-as0)+as0*rs))
W1
```

```python
EU=(prob_nat@W1.applyfunc(sp.log))[0]
EU
```

```python
as0_opt= sp.solve(sp.diff(EU,as0),as0)[0]
as0_opt
```

Note that the optimal fraction of wealth invested in risky asset does not depend on how wealthy we are. This is true for all preferences with constant relative risk aversion (like $u(W)=\log(W)$ or $u(W)=W^{\gamma}$)

### <font color='MediumVioletRed' style="font-size:20px"><b>Fundamental theorem of portfolio investment </b></font>

Suppose that the utility function for a given realization of terminal wealth $W$ is $u(W)$. In order to quantify risk aversion, it is very useful to consider coefficients of absolute and relative risk aversion. They are defined as follows:

$$A(W) = - \frac{u''(W)}{u'(W)}$$
$$R(W) = W A(W)$$

The more concave a utility function is, i.e. the bigger its curvature, the more risk averse the investor is. Note that for increasing concave utility functions both of these measures are positive.

**Fundamental theorem of portfolio optimization Part 1** states the following.

* Suppose $A(W)$ is an increasing function of $W$. In that case **the amount of money $w_{s0}$** that we optimally invest into the risky asset would *decrease* if we get *richer*. E.g. suppose I have 1,000 euros to invest and find that I would in that case opitmally invest 500 euros into the risky asset. Suppose now I get 10,000. According to this theorem, if my absolute coefficient of risk aversion increases in wealth I would end up investing now less than 500 euros. Not too realistic...

* Suppose $A(W)$ does not depend on $W$. In that case **the amount of money** that we optimally invest into the risky asset would stay the same if we get wealthier. In that case, suppose I optimally invest 500 euros in stock if I have 1,000 euros. If I all of a sudden have 10,000 euros to invest, I would still only invest 500 euros into the risky asset.

* Finally, suppose $A(W)$ decreases as a  function of $W$. Thus, we get less risk averse as we get wealthier in dollar terms. That would mean that we would optimally invest more money into risky asset than was the case when we had less money.

Consider $u(W)=\log(W)$. The absolute coefficient of risk aversion would in that case be

```python
W = sp.Symbol('W')
u = sp.log(W)
A = - sp.diff(u,W,2)/sp.diff(u,W)
A
```

Since A is a decreasing function of wealth, the wealthier we get the more dollars we are going to invest into the risky asset. This is intuitive. Moreover, we can check directly from our above calculations that this is true. If I double the wealth in the second formulation of the problem, $w_{s0}$ would increase.

**Fundamental theorem of portfolio optimization Part 2** states the following.

* Suppose $R(W)$ is an increasing function of $W$. In that case **the fraction of money $\alpha_{s}$** that we optimally invest into the risky asset would decrease if we get richer. E.g. suppose I have 1,000 euros to invest and find that I would in that case opitmally invest 50% into the risky asset. Suppose now I get 10,000. According to this theorem, if my relative coefficient of risk aversion increases in wealth I would end up investing now less than 50% into the risky asset.

* Suppose $R(W)$ does not depend on $W$. In that case **the fraction of money** that we optimally invest into the risky asset would stay the same whether we get wealthier or poorer. In that case, suppose I optimally invest 50% in stock if I have 1,000 euros. If I all of a sudden have 10000 euros to invest, I would still invest 50% of that new amount (i.e. 5,000 euros) into the risky asset.

* Finally, suppose $R(W)$ decreases as a  function of $W$. Thus, we get less risk averse in relative sense as we get wealthier. That would mean that we would optimally invest higher fraction of money into risky asset than was the case when we had less money.

Consider $u(W)=\log(W)$. The relative coefficient of risk aversion would be equal to

```python
W = sp.Symbol('W')
u = sp.log(W)
R = - W * sp.diff(u,W,2)/sp.diff(u,W)
R
```

Since the relative coefficient of risk aversion is a constant when investor has a log utility function, such investor would prefer to invest a constant percentage of his wealth (a constant fraction) into the risky asset however wealthy he becomes. We have observed this previously.

> Suppose we have utility function $u(W)$ which has constant relative risk aversion $R$. When we had 1,0000 dollars we optimally invested 3000 dollars into the risky asset. How much should we invest if our wealth increases to 10,0000?

The answer is quite simple. Since we assume constant relative risk aversion, we need to invest the same fraction of money into the risky asset, ie. 3,000.

## <font color='orange' style="font-size:25px"><b>Optimal portfolio problem with more than one risky asset</b></font>

Here I want to introduce yet another generalization of this portfolio optimization problem. Namely, in reality we may want to invest in portfolios which consists of more than one risky asset.

Suppose, for simplicity, that we can invest into two risky assets as well as into a riskless asset (straightforward to generalize to multiple risky assets). Risky assets are called a and b. The budget constraint at time 0 reads:

$$W_0 = B + n_{sa}\,p_{a0} + n_{sb}\,p_{b0}$$

Expressing from this the amount invested into the riskless asset $B$:

$$B = W_0 - n_{sa}\,p_{a0} - n_{sb}\,p_{b0}$$

At time 1, portfolio wealth would be:

$$W_1 = B (1 + r) + n_{sa}\,p_{a1} + n_{sb}\,p_{b1}$$

If we substitute expression for $B$ in equation above we will obtain the following result:

$$W_1 = (W_0 - n_{sa}\,p_{a0} - n_{sb}\,p_{b0})(1 + r) + n_{sa}\,p_{a1} + n_{sb}\,p_{b1}$$

We can rewrite the expression above as follows:

$$W_1=W_0 (1 + r) + n_a (p_{a1} - p_{a0} - p_{a0}\,r) + n_b (p_{b1} - p_{b0} - p_{b0}\,r)$$

Thus, final wealth differs from the amount we would have obtained by investing into the risk-free asset based on how much more (or less) each stock is making vis-a-vis the risk free asset.

If we label the amounts invested into a and b as: $w_{a0}= n_a\,p_{a0}$ and $w_{b0}= n_{b}\,p_{b0}$, and if we multiply quantities $n_a$ and $n_b$ with appropriate price at time $t=0$, we will obtain another expression for portfolio optimization (based on amount of money invested in risky assets):

$$W_1=W_0 (1 + r) + \frac{p_{a0}}{p_{a0}} n_a (p_{a1} - p_{a0} - p_{a0}\,r) + \frac{p_{b0}}{p_{b0}} n_b (p_{b1} - p_{b0} - p_{b0}\,r)$$

Which then yields:

$$W_1=W_0 (1 + r) + p_{a0}\,n_a \left(\frac{p_{a1} - p_{a0}}{p_{a0}} - r\right) + p_{b0}\,n_b \left(\frac{p_{b1} - p_{b0}}{p_{b0}} - r\right)$$

The wealth at time $W_1$ now reads:

$$W_1= W_0 (1 + r) + w_{a0} (r_a - r) + w_{b0}\,(r_b - r)$$

Finally, if we divide and multiply $w_{a0}$ and $w_{b0}$ with the initial wealth we woukld obtain the third possible expression for portfolio optimization (based on portfolio weights, or fractions of money invested in each asset):

$$W_1= W_0 (1 + r) + \frac{W_0}{W_0}\,w_{a0} (r_a - r) + \frac{W_0}{W_0}\,w_{b0}\,(r_b - r)$$

If we introduce two new variables:

$$\alpha_{a0}= \frac{w_{a0}}{W_0}\hspace{10mm} \text{and} \hspace{10mm} \alpha_{b0}= \frac{w_{b0}}{W_0},$$

Previous expression will be transformed into:

$$W_1= W_0 (1 + r + \alpha_{a0} (r_a - r) + \alpha_{b0}\,(r_b - r))$$

Which can then be rewriten as:

$$W_1= W_0 (1 + (1-\alpha_{a0}-\alpha_{b0})\,r + \alpha_{a0}\,r_a+ \alpha_{b0}\,r_b)$$

Where again we recognize weighted sum of returns or portfolio returns:

$$W_1= W_0 (1 + r_p)$$

In following couple of section we will discuss only optimization problems when we have two risky assets. However, case in which we have two risky and one riskless asset will be discussed in one of problems from today's problem set.

### <font color='MediumVioletRed' style="font-size:20px"><b>Just risky assets</b></font>

If we had just two risky assets and no riskless asset, expression for portfolio optimization based on number of stocks will be:

$$W_1=W_0+ n_a (p_{a1} - p_{a0}) + n_b (p_{b1} - p_{b0})$$

Expression for portfolio optimization based on amount of money invested in each stock becomes:

$$W_1= W_0 + w_{a0}\,r_a+ w_{b0}\,r_b$$

While expression for portion optimization based on weights (i.e. fractions) becomes:

$$W_1= W_0 (1 + \alpha_{a0}\,r_a+ \alpha_{b0}\,r_b)$$

#### <font color='MediumPurple' style="font-size:16px"><b>Numerical example</b></font>

> Suppose investor opportunity set is given by two risky stocks. At time 0, price of the first stock is 100 dollars while price of second stock is 90 dollars. Assume that there are only two possible states of nature. In up state first stock pays 120 dollars, while second stock pays 70 dollars.

> On the other hand, in down state, first stock pays 80 dollars, while second stock pays 110 dollars. Both states are equally probable. If investor has logarithmic utility function and he plans to invest 1000 dollars of his initial wealth, would be his/hers optimal portfolio?

Let us present payoffs with the asset tree:

```python
fd.tree([np.array([100]),np.array([[120,80]])],[np.array([90]),np.array([[70,110]])],space=0.25,fs=(500,500))
```

Let us define known parameters:

```python
W0=1000
pa=100
pb=90
prob_nat=sp.Matrix(1,2,[0.5,0.5])
```

Next, let us calculate net returns for both stocks:

```python
ra=np.array([120,80])/100-1
rb=np.array([70,110])/90-1
ra,rb
```

Next, let us compare this two stocks in terms of risk and reward. It is easy to see that the expected returns for both stocks are equal to zero:

```python
(prob_nat@sp.Matrix(ra))[0],round((prob_nat@sp.Matrix(rb))[0])
```

While their variances are:

```python
prob_nat@sp.Matrix((ra-0)**2),prob_nat@sp.Matrix((rb-0)**2)
```

Notice that the second stock is somewhat riskier than the first stock but has the same expected return. It would not be surprising that a risk-averse investor choses to invest more into the first stock.

Let us now get back to our problem. Since there are only two risky stocks, and since we know that the sum of all portfolio fractions has to sum up to 1, we can define fraction for stock b as $\alpha_{b0}=1-\alpha_{a0}$. Thus, wealth at time 1 is:

```python
aa0=sp.Symbol('a_a0')
W1=sp.Matrix(W0*(1+rb*(1-aa0)+aa0*ra))
W1
```

Now we need to define expected utility function (remember that our investor has logarithmic utility function):

```python
EU =(prob_nat@W1.applyfunc(sp.log))[0]
EU
```

The optimal fractions invested in two stocks is obtained by solving for the first order condition:

```python
aa0_opt = sp.solve(EU.diff(aa0),aa0)[0]
aa0_opt,1-aa0_opt
```

Thus, indeed, we would invest more into the less risky stock. However, assume that probabilities have changed in favor of the down state in which the second (previously riskier) stock has higher payoffs. For example assume following probabilities:

```python
prob_nat2= sp.Matrix(1,2,[0.45,0.55])
```

So, new expected utility function will be:

```python
EU= (prob_nat2@W1.applyfunc(sp.log))[0]
EU
```

And, finally, new optimal weights would be:

```python
aa0_opt=sp.solve(EU.diff(aa0),aa0)[0]
aa0_opt, 1-aa0_opt
```

And as you can see, in this case weight given to the second stock is much higher!
